## Recursive Approach
We use a base case: For 0th step
```python
return x_start """first step"""
```
We use the following base case: For 1st step
```python
return y_axis[n]
```
### Observation
When we use `return y_axis[0]` we somehow obtain an increasing graph over multiple iterations. This could perhaps be explained by the low degree of randomisation


````python
import matplotlib.pyplot as plt
import numpy as np
import random
````

````python
iter=1000
x_axis=np.random.randint(2,size=iter)
y_axis=x_axis*2 -1
"""outptus 1 if 2, or -1 if 0"""
x_start=0
"""function called with number iter later"""
def random_walk(n):
  if n==0:
    return x_start
  if n==1:
    return y_axis[n]
  else:
    x=random_walk(n-1)+y_axis[n]
    return x
"""if we call with n=0, then we have no plot points to generate. """
"""if n = 1, then we return last al of b_steps"""
"""we call this function recursively, so we define our base cases"""
l_n=[i for i in range(iter)]
l_rn=[random_walk(i) for i in range(iter)]
plt.plot(l_n,l_rn,label='random walk')
plt.xlabel('number of iteratiojs')
plt.ylabel('random number generated')
````

## Iterative Approach
### Using Cumulative Sum
Instead of using base cases for 0 and 1, we can now also initialise each value of the y_axis array based on the previous element. We use only one base case, that is, the starting point. This is a much more efficient computation.


````python

x_axis=np.random.randint(2,size=40)
y_axis=x_axis * 2 - 1
y_cum = y_axis

"""outptus 1 if 2, or -1 if 0"""
x_start = 0

for x in range(1, 40):
    y_cum[x] += y_cum[x-1] + y_axis[x]
  
"""if we call with n=0, then we have no plot points to generate. """
"""if n = 1, then we return last al of b_steps"""
"""we call this function recursively, so we define our base cases"""
l_n=[i for i in range(40)]
    
plt.plot(l_n,y_cum,label='random walk')
plt.xlabel('number of iterations')
plt.ylabel('random number generated')
````
